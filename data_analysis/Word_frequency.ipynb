```python
# Lấy văn bản từ file Luat.txt
file = open("../dataset/Luat.txt", "r", encoding="utf-8")
content = file.read()
file.close()
```

# Từ đơn

```python
from collections import Counter
split_it = content.split()
Counter = Counter(split_it)
most_occur = Counter.most_common(20)

for word, count in most_occur:
    print(f"{word}: {count}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.DataFrame(most_occur, columns=['Word', 'Word frequency'])
colors = sns.color_palette("coolwarm", len(df))
sns.barplot(x='Word frequency', y='Word',
            data=df, orient="h",
            palette=colors)
plt.title("Most frequent words")
plt.show()
```

# Cặp từ

```python
from collections import Counter

bigrams = zip(split_it, split_it[1:])
counts = Counter(bigrams)
most_occur_bigrams = counts.most_common(20)

for bigram, count in most_occur_bigrams:
    print(f"{bigram[0]} {bigram[1]}: {count}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from collections import Counter

# Sample data
split_it = content.split()
bigrams = zip(split_it, split_it[1:])
counts = Counter(bigrams)
most_occur_bigrams = counts.most_common(20)

# Prepare data for plotting
bigram_strings = [' '.join(bigram) for bigram, count in most_occur_bigrams]
bigram_counts = [count for bigram, count in most_occur_bigrams]
df = pd.DataFrame(list(zip(bigram_strings, bigram_counts)), columns=['Bigram', 'Frequency'])

# Custom color palette from blue to red
colors = sns.color_palette("coolwarm", len(df))

# Plotting
sns.barplot(x='Frequency', y='Bigram', data=df, orient="h", palette=colors)
plt.title("Most frequent bigrams")
plt.show()
```
